```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_json('states_daily.json')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset['date'] = pd.to_datetime(dataset['date'])
```

```python
plt.barh(dataset['status'], dataset['mh'])
```

```python
plt.figure(figsize=(10,6))
plt.barh(dataset['status'], dataset['dl'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot_date(dataset['date'], dataset['dl'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(dataset['date'], dataset['dl'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
confirmed = dataset[dataset['status'] == 'Confirmed']
recovered = dataset[dataset['status'] == 'Recovered']
```

```python
plt.figure(figsize=(10,6))
plt.plot(confirmed['date'], confirmed['dl'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(confirmed['date'], confirmed['mh'],label = 'MH')
plt.plot(confirmed['date'], confirmed['dl'], label = 'DL')
plt.xticks(rotation='vertical')
plt.legend()
plt.title('COVID-19 Cases in Maharastra and Delhi')
plt.show()
```
